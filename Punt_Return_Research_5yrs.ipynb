```python
import pandas as pd
import numpy as np
import nfl_data_py as nfl
import matplotlib.pyplot as plt
full_data = nfl.import_pbp_data([2017, 2018, 2019, 2020, 2021], downcast=True, cache=False, alt_path=None)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
newData = full_data[['game_date', 'play_id', 'play_type_nfl', 'punt_inside_twenty', 'game_id', 'possession_team', 'home_team', 'away_team', 'home_score', 'away_score', 'total_home_score', 'total_away_score', 'wp', 'game_half', 'game_seconds_remaining', 'play_type', 'def_wp', 'return_yards', 'return_touchdown']]
newData = newData.rename(columns={"wp": "win_percentage", "def_wp": "defense_win_percentage"})
puntData = newData[newData["return_yards"] >= 20.0]
puntData = puntData[puntData["play_type_nfl"] == 'PUNT']
puntData
```
