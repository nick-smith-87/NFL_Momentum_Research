```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df15 = pd.read_csv("./Data/Positional Spending - Pos_Spend_15.csv")
df16 = pd.read_csv("./Data/Positional Spending - Pos_Spend_16.csv")
df17 = pd.read_csv("./Data/Positional Spending - Pos_Spend_17.csv")
df18 = pd.read_csv("./Data/Positional Spending - Pos_Spend_18.csv")
df19 = pd.read_csv("./Data/Positional Spending - Pos_Spend_19.csv")
df20 = pd.read_csv("./Data/Positional Spending - Pos_Spend_20.csv")
df21 = pd.read_csv("./Data/Positional Spending - Pos_Spend_21.csv")
df22 = pd.read_csv("./Data/Positional Spending - Pos_Spend_22.csv")

df = pd.concat([df15, df16, df17, df18, df19, df20, df21, df22])
df
```

```python
qb_money = df.groupby('Year')['QB Adjusted'].mean()
rb_money = df.groupby('Year')['RB Adjusted'].mean()
wr_money = df.groupby('Year')['WR Adjusted'].mean()
te_money = df.groupby('Year')['TE Adjusted'].mean()
ol_money = df.groupby('Year')['OL Adjusted'].mean()
dl_money = df.groupby('Year')['DL Adjusted'].mean()
lb_money = df.groupby('Year')['LB Adjusted'].mean()
db_money = df.groupby('Year')['DB Adjusted'].mean()
st_money = df.groupby('Year')['Special Teams Adjusted'].mean()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(qb_money.index, qb_money, label='QB')
plt.plot(rb_money.index, rb_money, label='RB')
plt.plot(wr_money.index, wr_money, label='WR')
plt.plot(te_money.index, te_money, label='TE')
plt.plot(ol_money.index, ol_money, label='OL')
plt.plot(dl_money.index, dl_money, label='DL')
plt.plot(lb_money.index, lb_money, label='LB')
plt.plot(db_money.index, db_money, label='DB')
plt.plot(st_money.index, st_money, label='Special Teams')

plt.xlabel('Year')
plt.ylabel('Average Adjusted Salary')
plt.title('Average Adjusted Salary by Position Over Years')
plt.legend()
plt.grid(True)
plt.show()
```
