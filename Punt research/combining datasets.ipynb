```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
X, y = load_iris(return_X_y=True)
```

```python
clf = LogisticRegression(random_state=0).fit(X, y)
```

```python
clf.predict(X[:2, :])
```

```python
clf.predict_proba(X[:2, :])
```

```python
clf.score(X, y)
```
