```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Sdf = pd.read_csv("./data/Sauce Gardner 2022 - Sheet1.csv")
Tdf = pd.read_csv("./data/Tariq Woolen 2022 - Sheet1.csv")
```

```python
Sdf["player"] = "Sauce Gardner"
Tdf["player"] = "Tariq Woolen"
Sdf['Cumulative Yards'] = Sdf['Yards'].cumsum()
Tdf['Cumulative Yards'] = Tdf['Yards'].cumsum()
```

```python
combined = pd.concat([Sdf, Tdf])

fig, ax = plt.subplots(figsize=(10,6))
sns.lineplot(data=combined, x="Week", y="Cumulative Yards", hue="player", ax=ax)
ax.set_xlabel("Week")
ax.set_ylabel("Cumulative Yards")
ax.set_title("Yards allowed throughout the season")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate interception rate for each player
S_interception_rate = Sdf['INTS'].sum() / Sdf['Targets'].sum()
T_interception_rate = Tdf['INTS'].sum() / Tdf['Targets'].sum()

# Calculate completion percentage allowed for each player
S_completion_pct = Sdf['Receptions'] / Sdf['Targets']
T_completion_pct = Tdf['Receptions'] / Tdf['Targets']

# Calculate yards per target allowed for each player
S_yards_per_target = Sdf['Yards'] / Sdf['Targets']
T_yards_per_target = Tdf['Yards'] / Tdf['Targets']

# Create bar chart to compare interception rates
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(['Player 1', 'Player 2'], [S_interception_rate, T_interception_rate])
ax.set_title('Interception Rate Comparison')
ax.set_xlabel('Player')
ax.set_ylabel('Interception Rate')

# Create line chart to compare completion percentages
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(S_completion_pct, label='Sauce Gardner')
ax.plot(T_completion_pct, label='Tariq Woolen')
ax.set_title('Completion Percentage Comparison')
ax.set_xlabel('Week')
ax.set_ylabel('Completion Percentage')
ax.legend()

# Create line chart to compare yards per target allowed
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(S_yards_per_target, label='Player 1')
ax.plot(T_yards_per_target, label='Player 2')
ax.set_title('Yards per Target Allowed Comparison')
ax.set_xlabel('Week')
ax.set_ylabel('Yards per Target')
ax.legend()

plt.show()
```
