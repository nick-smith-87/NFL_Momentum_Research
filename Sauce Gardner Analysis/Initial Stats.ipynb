```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./data/defense_coverage_summary.csv")
data.head()
```

```python
df = data
pd.set_option('display.max_columns', None)
df = df[df.position == "CB"]
cb_df = df.loc[df['player'].isin(["Sauce Gardner", "Patrick Surtain II", "Patrick Peterson", "Jaire Alexander", "Denzel Ward"])]
cb_df
```

```python
colors = ['yellow', 'green', 'orange', 'blue', 'green'] 
plt.bar(cb_df.player, cb_df.yards, color = colors)
plt.xlabel("NFL Corners")
plt.ylabel("Yards Allowed")
plt.title("NFL 2022 Season Top NFL Corners")
plt.xticks(rotation=45)

plt.show()
```
